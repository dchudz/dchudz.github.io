```julia
using Gadfly
```

```julia
griddf = DataFrame(x_min=[1,3], x_max=[2,4], y_min=[3,4], y_max=[4,5], col=[1,2])
pointsdf = DataFrame(x=[2:3], y=[4:5])

plot(
layer(pointsdf, x=:x, y=:y, Geom.point),
layer(griddf, x_min=:x_min, x_max=:x_max, y_min=:y_min, y_max=:y_max, color=:col, Geom.rectbin)
)
```
